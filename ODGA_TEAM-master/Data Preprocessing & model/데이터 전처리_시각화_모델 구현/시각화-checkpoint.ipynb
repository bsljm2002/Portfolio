```python
from scipy.signal import medfilt
import numpy as np
import scipy as sp


def median(signal_x, signal_y, signal_z):
    array_x = np.array(signal_x)
    array_y = np.array(signal_y)
    array_z = np.array(signal_z)
    med_filtered_x=sp.signal.medfilt(array_x, kernel_size=3)
    med_filtered_y=sp.signal.medfilt(array_y, kernel_size=3)
    med_filtered_z=sp.signal.medfilt(array_z, kernel_size=3)
    return med_filtered_x, med_filtered_y, med_filtered_z
```
